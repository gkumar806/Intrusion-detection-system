```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

1. Load Dataset

```python
data=pd.read_csv("data.csv")
```

2. Encode categorical columns


```python
label_encoders = {}
for col in data.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le
```

3. Split features and target

```python
X = data.drop('class', axis=1)
y = data['class']
```

4. Train-test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

5. Train RandomForest Model

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

6. Feature importance

```python
importances = model.feature_importances_
feature_names = X.columns
```

7. Create a Dataframe for visulization

```python
feat_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feat_importance_df = feat_importance_df.sort_values(by='Importance', ascending=False)
```

8. Plot top 10 Features

```python
plt.figure(figsize=(10,6))
plt.barh(feat_importance_df['Feature'].head(10)[::-1], feat_importance_df['Importance'].head(10)[::-1], color='skyblue')
plt.xlabel('Importance')
plt.title('Top 10 Features for Attack Detection')
plt.show()
```

9. Print top 10 features

```python
print("üîπ Top 10 Important Features:\n", feat_importance_df.head(10))
```
